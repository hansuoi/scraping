```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
urlName = "https://www.imsdb.com/scripts/La-La-Land.html"
url = requests.get(urlName)
soup = BeautifulSoup(url.content, "html.parser")
```

```python
# 先頭のBタグの中身(= タイトル)をscriptに格納
script = soup.find("b")


b = soup.findAll("b")
i = 1

title = re.findall('\s{26}([^\r]*)\r\n', script.text)[0]
print('title  : ', title)

script = script.next_sibling
writer = re.findall('\s*Written\sby\s*([^\r]*)', script.string)[0]
print('writer : ', writer)

print('--------------------')


char_flag = False
while script.next_sibling != None:
    script = script.next_sibling
    if script.string == b[i].text:
        i += 1
        if re.findall('^\s([^\r])*', script.string) != []:
            print('char   : ', script.string.strip())
            char_flag = True
        elif re.findall('([^\r])*', script.string) != []:
            print('scene  : ', script.string.strip())
    else:
        s = script.string
        if char_flag:
            print('lines  : ', end='')
            j = 0
            while s[0] == ' ':
                s = s[1:]
            while len(s) > j+3  and  s[j] + s[j+1] + s[j+2] + s[j+3] != '\r\n\r\n':
                print(s[j].replace('\n', ' '), end='')
                j += 1
            print('\n')
            char_flag = False
        else:
            print('writing: ', script.string.strip().replace('\r\n', ' '), end='\n\n')
```
